```python
%matplotlib inline
from matplotlib import pyplot as plt
```

```python
plt.plot([1,2,3], [2,5,7])
plt.title("My Graph")
plt.xlabel("X - axis")
plt.ylabel("Y - axis")
plt.show()
```

```python
from matplotlib import style

style.use('ggplot')
x = [1,1,3,3]
y = [1,5,1,5]
x1 = [5,5,7,7,5]
y1 = [1,5,5,3,3]
plt.plot(x,y, label='first', linewidth=3)
plt.plot(x1,y1, label='second', linewidth=3)
plt.legend()
plt.show()
```
